```python
import pandas as pd
```

```python
school_data="Resources/schools_complete.csv"
student_data="Resources/students_complete.csv"
```

```python
school_data_df=pd.read_csv(school_data)
student_data_df=pd.read_csv(student_data)

total_data_df=pd.merge(student_data_df,school_data_df, how="left", on=["school_name"] )
total_data_df.head()
```

```python
schools=pd.unique(total_data_df["school_name"])
```

```python
number_of_schools=len(schools)
```

```python
amount_of_students=len(total_data_df)
```

```python
budget=pd.unique(total_data_df["budget"])
total_budget=sum(budget)
```

```python
average_reading_score=total_data_df["reading_score"].mean()
```

```python
average_math_score=total_data_df["math_score"].mean()
```

```python
only_passing_math=total_data_df.loc[total_data_df["math_score"]>=70]
number_passing_math=len(only_passing_math)
percent_passing_math=(number_passing_math/amount_of_students)*100
```

```python
only_passing_reading=total_data_df.loc[total_data_df["reading_score"]>=70]
number_passing_reading=len(only_passing_reading)
percent_passing_reading=(number_passing_reading/amount_of_students)*100
```

```python
passing_both=total_data_df.loc[(total_data_df["reading_score"]>=70) & (total_data_df["math_score"]>=70)]
number_passing_both=len(passing_both)
percent_passing_both=(number_passing_both/amount_of_students)*100
```

```python
district_summary={
    "Total Schools": number_of_schools,
    "Total Students": amount_of_students,
    "Total Budget": total_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_passing_both
}
district_summary_df=pd.DataFrame(district_summary, index=[0])
district_summary_df.head()
```

```python
school_counts = total_data_df["school_name"].value_counts()
```

```python
grouped_school_df= total_data_df.groupby(['school_name'])
grouped_school_df.count().head()
```

```python
budget=grouped_school_df["budget"].sum()/school_counts
```

```python
budget_per_student=budget/school_counts
```

```python
math_score=(grouped_school_df["math_score"].sum()/school_counts)
```

```python
reading_score=(grouped_school_df["reading_score"].sum()/school_counts)
```

```python
students_passing_math_per_school = only_passing_math.groupby(["school_name"]).count()["student_name"]
percent_passing_math_per_school=(students_passing_math_per_school/school_counts)*100
```

```python
students_passing_reading_per_school = only_passing_reading.groupby(["school_name"]).count()["student_name"]
percent_passing_reading_per_school=(students_passing_reading_per_school/school_counts)*100
```

```python
students_passing_both_per_school = passing_both.groupby(["school_name"]).count()["student_name"]
percent_passing_both_per_school=(students_passing_both_per_school/school_counts)*100
```

```python
school_type = school_data_df.set_index(["school_name"])["type"]
```

```python
school_summary={
    "School Type": school_type, 
    "Total Students": school_counts, 
    "Total School Budget": budget, 
    "Per Student Budget": budget_per_student, 
    "Average Math Score": math_score,
    "Average Reading Score": reading_score,
    "% Passing Math": percent_passing_math_per_school,
    "% Passing Reading": percent_passing_reading_per_school,
    "% Overall Passing": percent_passing_both_per_school
}
school_summary_df=pd.DataFrame(school_summary)
school_summary_df.head()
```

```python
top_performing=school_summary_df.sort_values("% Overall Passing", ascending=False)
top_performing.head()
```

```python
bottom_performing=school_summary_df.sort_values("% Overall Passing")
bottom_performing.head()
```

```python
grade_nine=total_data_df.loc[total_data_df["grade"]=="9th",:]
grade_nine_grouped=grade_nine.groupby(['school_name'])
grade9_counts = grade_nine["school_name"].value_counts()
math_score_nine=(grade_nine_grouped["math_score"].sum()/grade9_counts)
reading_score_nine=(grade_nine_grouped["reading_score"].sum()/grade9_counts)
```

```python
grade_10=total_data_df.loc[total_data_df["grade"]=="10th",:]
grade_10_grouped=grade_10.groupby(['school_name'])
grade10_counts = grade_10["school_name"].value_counts()
math_score_10=(grade_10_grouped["math_score"].sum()/grade10_counts)
reading_score_10=(grade_10_grouped["reading_score"].sum()/grade10_counts)
```

```python
grade_11=total_data_df.loc[total_data_df["grade"]=="11th",:]
grade_11_grouped=grade_11.groupby(['school_name'])
grade11_counts = grade_11["school_name"].value_counts()
math_score_11=(grade_11_grouped["math_score"].sum()/grade11_counts)
reading_score_11=(grade_11_grouped["reading_score"].sum()/grade11_counts)
```

```python
grade_12=total_data_df.loc[total_data_df["grade"]=="12th",:]
grade_12_grouped=grade_12.groupby(['school_name'])
grade12_counts = grade_12["school_name"].value_counts()
math_score_12=(grade_12_grouped["math_score"].sum()/grade12_counts)
reading_score_12=(grade_12_grouped["reading_score"].sum()/grade12_counts)
```

```python
math_by_grade={
    "9th": math_score_nine,
    "10th": math_score_10,
    "11th": math_score_11,
    "12th": math_score_12
}
math_by_grade_df=pd.DataFrame(math_by_grade)
math_by_grade_df.head(15)
```

```python
reading_by_grade={
    "9th": reading_score_nine,
    "10th": reading_score_10,
    "11th": reading_score_11,
    "12th": reading_score_12
}
reading_by_grade_df=pd.DataFrame(reading_by_grade)
reading_by_grade_df.head(15)
```

```python
bins=[0,585,630,645,675]
names=["<$585", "$585-$629", "$630-$644", "$645-674"]
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"],bins,labels = names)
```

```python
avg_math_score_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
avg_reading_score_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
students_passing_math_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"] 
students_passing_reading_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_pass_rate_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
scores_based_spending = {"Average Math Score":avg_math_score_spending,
                                     "Average Reading Score":avg_reading_score_spending,
                                     "% Passing Math":students_passing_math_spending,
                                     "%Passing Reading":students_passing_reading_spending ,
                                     "Overall Passing rate":overall_pass_rate_spending}
scores_bsaed_spending_df=pd.DataFrame(scores_based_spending)
scores_bsaed_spending_df
```
